### 1.3.2 Ver se√±ales de relevancia inicial
Para num√©ricos, correlaci√≥n con Churn binario.

```python
# Convertimos Charges.Total a num√©rico para evaluar
hackathon_df['MonthlyRevenue'] = pd.to_numeric(hackathon_df['MonthlyRevenue'], errors='coerce')

# Crear versi√≥n binaria de Churn
churn_bin = hackathon_df['Churn'].map({'Yes': 1, 'No': 0})

# Correlaciones para num√©ricos
num_cols = [
    col for col in hackathon_df.select_dtypes(include=['int64', 'float64']).columns
    if col != 'CustomerID'
]

# Calcular correlaciones
corr_churn = hackathon_df[num_cols].corrwith(churn_bin).sort_values(ascending=False)
```

```python
# Tasa de churn por categor√≠a para algunas columnas
# Variables a analizar
cols_to_analyze = [
    'CurrentEquipmentDays',
    'PercChangeMinutes',
    'MonthlyMinutes',
    'MonthsInService',
    'ActiveSubs'
]

# Crear bins para variables num√©ricas
for col in cols_to_analyze:
    hackathon_df[f'{col}_bin'] = pd.qcut(
        hackathon_df[col],
        q=4,
        duplicates='drop'
    )

# Tasa de churn por categor√≠a
for col in cols_to_analyze:
    print(
        f"\n{col} (binned):\n",
        hackathon_df
        .groupby(f'{col}_bin', observed=True)['Churn']
        .value_counts(normalize=True)
        .unstack()
    )
```

## 3. Manejo de inconsistencias

Aplicamos transformaciones para garantizar datos consistentes y listos para an√°lisis:



### 3.1 Variables Num√©ricas
- `MonthlyRevenue` y `TotalRecurringCharge`: ya no tienen strings.
- Convertir `floats` que representan enteros a `int` (ej. 14.0 ‚Üí 14) si corresponde.
- Reemplazar valores nulos (`NaN`) por 0 solo si tiene sentido. (ej. cargos mensuales).

```python
num_cols = ['MonthlyRevenue', 'MonthlyMinutes', 'TotalRecurringCharge',
            'DirectorAssistedCalls', 'OverageMinutes', 'RoamingCalls']
for col in num_cols:
    hackathon_df[col] = hackathon_df[col].fillna(0).astype(int)
```

### 3.2 Variables categ√≥ricas
- Normalizar may√∫sculas/min√∫sculas y eliminar espacios extra
- Reemplazar `'none'` o `'unknown'` por `NaN` en general (si aplica), excepto en `ServiceArea`


```python
# Normalizar may√∫sculas/min√∫sculas y eliminar espacios extra
cat_cols = hackathon_df.select_dtypes(include='object').columns
for col in cat_cols:
    hackathon_df[col] = hackathon_df[col].str.strip()

# Valores inconsistentes a pd.NA (excepto ServiceArea que tendr√° UNKNWN000)
hackathon_df.loc[:, cat_cols] = hackathon_df.loc[:, cat_cols].replace({'unknown': pd.NA, 'none': pd.NA})

```

```python
hackathon_df['Churn'] = hackathon_df['Churn'].fillna(0)  # Asumimos que 'NaN' es 0 para Churn
```

### 5.2 Lista con las columnas que contienen "Yes" o "No"


```python
binary_cols = [
    'Churn', 'ChildrenInHH', 'HandsetRefurbished', 'HandsetWebCapable',
    'TruckOwner', 'RVOwner', 'BuysViaMailOrder', 'RespondsToMailOffers',
    'OptOutMailings', 'NonUSTravel', 'OwnsComputer', 'HasCreditCard',
    'NewCellphoneUser', 'NotNewCellphoneUser', 'OwnsMotorcycle',
    'MadeCallToRetentionTeam'
]
```

### 5.3 Verificaci√≥n de valores err√≥neos en las columnas binarias ('Maybe', 'Unknown', etc.)


```python
# Limpiar cualquier valor extra√±o (como 'Maybe', 'Unknown')
for col in binary_cols:
    hackathon_df[col] = hackathon_df[col].replace(['Maybe', 'Unknown'], pd.NA)
```

### 5.4 Transformar las columnas binarias (Yes/No a 1/0)


```python
for col in binary_cols:
    # Comprobamos los valores √∫nicos antes de realizar el mapeo
    print(f"Valores √∫nicos en {col}: {hackathon_df[col].unique()}")

    # Si hay valores no v√°lidos (como NaN despu√©s de la limpieza), =  0
    hackathon_df[col + '_bin'] = hackathon_df[col].map({'Yes': 1, 'No': 0}).fillna(0)
```

### 5.5 Comprobaci√≥n de valores err√≥neos en las columnas binned


```python
# Lista de columnas binned que necesitan ser verificadas
binned_cols = [
    'CurrentEquipmentDays_bin', 'PercChangeMinutes_bin', 'MonthlyMinutes_bin',
    'MonthsInService_bin', 'ActiveSubs_bin'
]

# Verificar si hay NaN en las columnas binned
for col in binned_cols:
    print(f"Valores √∫nicos en {col}: {hackathon_df[col].unique()}")

# Convertir las columnas binned a valores medios
for col in binned_cols:
    hackathon_df[col] = hackathon_df[col].apply(lambda x: (x.left + x.right) / 2)

# Verificar los cambios
print(hackathon_df[binned_cols].head())
print(hackathon_df[binned_cols].dtypes)
```

### 5.6 Rellenar valores NaN en columnas num√©ricas con la media


```python
numeric_cols = ['PercChangeMinutes', 'PercChangeRevenues', 'Handsets', 'HandsetModels',
                'CurrentEquipmentDays']
for col in numeric_cols:
    hackathon_df[col] = hackathon_df[col].astype(float)
    hackathon_df[col] = hackathon_df[col].fillna(hackathon_df[col].mean())

# Rellenar valores NaN en columnas binarias/categ√≥ricas con el valor m√°s frecuente (modo)
binary_cols = ['CurrentEquipmentDays_bin', 'PercChangeMinutes_bin', 'MonthlyMinutes_bin']
for col in binary_cols:
    hackathon_df[col] = hackathon_df[col].fillna(hackathon_df[col].mode()[0])

# Rellenar valores NaN en AgeHH1 y AgeHH2 con la mediana (si son num√©ricas)
hackathon_df['AgeHH1'] = hackathon_df['AgeHH1'].fillna(hackathon_df['AgeHH1'].median())
hackathon_df['AgeHH2'] = hackathon_df['AgeHH2'].fillna(hackathon_df['AgeHH2'].median())

# Verificar si hay m√°s valores NaN
print(hackathon_df.isna().sum())
```

```python
import os
TARGET_COL = 'Cancelaci√≥n_bin'
DAILY_REVENUE_COL = 'Ingreso_Diario'

# 1. Dataset 1 (backup / EDA)
df_raw_clean = hackathon_df.copy()


# 2. Dataset 2 (MVP PRINCIPAL)
# Columnas binarias (_bin)
binary_model_cols = [
    col for col in hackathon_df.columns
    if col.endswith('_bin') and col != TARGET_COL
]

# Columnas num√©ricas reales (sin bins)
numeric_model_cols = [
    col for col in hackathon_df.columns
    if hackathon_df[col].dtype in ['int64', 'float64']
    and not col.endswith('_bin')
    and col not in [TARGET_COL]
]

# Columnas finales del modelo base
model_base_cols = list(
    set(
        numeric_model_cols +
        binary_model_cols +
        [DAILY_REVENUE_COL, TARGET_COL]
    )
)

# Filtrar solo columnas existentes (anti-errores)
model_base_cols = [c for c in model_base_cols if c in hackathon_df.columns]

df_model_base = hackathon_df[model_base_cols].copy()

# 3. Dataset 3 (experimento)
binned_cols = [
    'D√≠as_Equipo_Actual_bin',
    'Cambio_Perc_Minutos_bin',
    'Minutos_Mensuales_bin',
    'Meses_En_Servicio_bin',
    'Subs_Activos_bin'
]

binary_cols = [
    col for col in hackathon_df.columns
    if col.endswith('_bin')
]

model_binned_cols = list(
    set(
        binned_cols +
        binary_cols +
        [DAILY_REVENUE_COL, TARGET_COL]
    )
)

# Seguridad
model_binned_cols = [c for c in model_binned_cols if c in hackathon_df.columns]

df_model_binned = hackathon_df[model_binned_cols].copy()
```
