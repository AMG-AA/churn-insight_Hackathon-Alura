#2️⃣ Preparación segura: separar tipos de variables (SIN errores)

Este código detecta automáticamente qué columnas son numéricas y cuáles categóricas.

✔ No falla

✔ No importa cuántas columnas tengas

✔ Funciona en Colab

📝 Texto profesional para tu notebook

La variable objetivo abandono fue transformada a formato binario (1 = Yes, 0 = No) para permitir el análisis estadístico y el entrenamiento de modelos de clasificación.

```python
# Variable objetivo
target = 'abandono'
# Conversión segura de Yes / No a 1 / 0
df[target] = df[target].map({'Yes': 1, 'No': 0})
# Variables numéricas (excluyendo el target)
variables_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
variables_numericas = [col for col in variables_numericas if col != target]

# Variables categóricas
variables_categoricas = df.select_dtypes(include=['object']).columns.tolist()
```

#3️⃣ Relevancia de variables numéricas (correlación)

📌 Interpretación:

valores cercanos a 1 o -1 → más relevantes

valores cerca de 0 → poco impacto

```python
correlacion = df[variables_numericas + [target]].corr()[target].sort_values(ascending=False)

correlacion
```

#4️⃣ Relevancia de variables categóricas (tasa de churn)

Este código NO genera error, incluso con texto:

📌 Esto responde preguntas como:

¿Qué área_servicio tiene más abandono?

¿Qué tipo_plan cancela más?

```python
for col in variables_categoricas:
    print(f'\nTasa de abandono por {col}')
    display(
        df.groupby(col)[target]
          .mean()
          .sort_values(ascending=False)
    )
```

#5️⃣ Importancia usando un modelo (forma más fuerte)

Usamos Random Forest porque:

maneja bien datos categóricos codificados

no exige normalidad

es ideal para churn

🔹 Paso 1: Codificar variables categóricas

```python
from sklearn.preprocessing import LabelEncoder

df_modelo = df.copy()

le = LabelEncoder()
for col in variables_categoricas:
    df_modelo[col] = le.fit_transform(df_modelo[col])
```

🔹 Paso 2: Entrenar modelo

```python
from sklearn.ensemble import RandomForestClassifier

X = df_modelo.drop(columns=target)
y = df_modelo[target]

modelo = RandomForestClassifier(random_state=42)
modelo.fit(X, y)
```

🔹 Paso 3: Importancia de variables

📌 Este ranking es el más usado en hackathons y proyectos reales.

---



```python
import pandas as pd

importancia = pd.DataFrame({
    'variable': X.columns,
    'importancia': modelo.feature_importances_
}).sort_values(by='importancia', ascending=False)

importancia
```

#2️⃣ Código para obtener SOLO las variables con correlación > 0.05
📌 Resultado:

Lista clara de variables numéricas relevantes

Ordenadas de más a menos importantes

```python
# Calcular correlación absoluta con el target
correlacion = (
    df[variables_numericas + [target]]
    .corr()[target]
    .abs()
)

# Filtrar variables con correlación > 0.05 (excluyendo el target)
variables_correlacionadas = correlacion[
    (correlacion > 0.05) & (correlacion.index != target)
]

variables_correlacionadas.sort_values(ascending=False)
```

```python
#Verlas como lista (para selección final)
variables_numericas_relevantes = variables_correlacionadas.index.tolist()
variables_numericas_relevantes
```
